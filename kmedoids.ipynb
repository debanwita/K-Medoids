```python
class Kmedoid(object):
    def __init__(self,c):
        self.clusters = c
        "self.X = np.array(([2, 6], [3, 4], [3, 8], [4, 7], [6, 2], [6, 4], [7, 3], [7, 4], [8, 5], [7, 6]), dtype=float) # (2,6),(3,4),(3,8),(4.7) in one cluster and others in another cluster "
        
        self.X = np.array(([3 , 7] ,
            [5, 5] ,
            [6, 8],
            [7, 4],
            [5, 9]), 
            dtype=float) 
        self.attributes = len(self.X[0])
    
    def kmedoid_clustering(self):
        cluster_center = np.ndarray(shape=(self.clusters, self.attributes), dtype=float)
        cluster_center_prev = np.ndarray(shape=(self.clusters, self.attributes), dtype=float)
        for i in range(self.clusters):
            cluster_center[i] = self.X[i]
        #print(cluster_center)
        
        
        cluster_array = np.ndarray(shape=(self.clusters, len(self.X)), dtype=int)
      
        epsilon = 0.01
        while True:
            indices = np.zeros(self.clusters, dtype=int)
            
            for i in range(len(self.X)):
                
                #calculate distances b/w each point and cluster centers
                min_dist = 9999

                for j in range(self.clusters):
                    #d= math.pow((self.X[i][0] - cluster_center[j][0]), 2) + math.pow((self.X[i][1] - cluster_center[j][1]), 2)
                    d = 0.0
                    for g in range(self.attributes):
                        d = d + math.pow((self.X[i, g] - cluster_center[j, g]), 2)
                        
                    if(min_dist > d):
                        min_dist = d
                        min_cluster = j 
                    #print(d)

                #print(min_cluster)
                k = indices[min_cluster]
                cluster_array[min_cluster, k] = i

                #print(k)
                print("point ", cluster_array[min_cluster, k] , self.X[i], " in cluster ", min_cluster)
                k = k + 1
                indices[min_cluster] = k
          
                #print(indices)

            for n in range(self.clusters):
                number_of_points = indices[n] 
                #print(number_of_points)
                #print("---------")

                #xvalue = 0.0
                #yvalue = 0.0
                values = np.zeros((self.attributes), dtype=float)
                for m in range(number_of_points):
                    p = cluster_array[n, m]
                    #print(self.X[p][0], self.X[p][1])
                    
                    #xvalue =  xvalue + self.X[p][0] 
                    #yvalue =  yvalue + self.X[p][1]
                    
                    for g in range(self.attributes):
                        values[g] = values[g] + self.X[p][g]
                    "print(values[g])"


                #print(xvalue , yvalue , number_of_points)
                #xvalue = xvalue / number_of_points
                #yvalue = yvalue / number_of_points
                
                for g in range(self.attributes):
                        values[g] = values[g] / number_of_points
                        

                #calcuate distance b/w every points in the cluster and the cluster center
                min2=9999
                for l in range(number_of_points):
                    p = cluster_array[n, m]
                    dist = 0.0
                    for g in range(self.attributes):
                        dist = dist + math.pow((self.X[p, g] - values[g]), 2)
                    #dist = math.pow((self.X[p][0] - xvalue), 2) + math.pow((self.X[p][1] - yvalue), 2)
                    if(min2 > dist):
                        min2 = dist
                        c_center = p
                    
                #cluster_center_prev[n, 0] = cluster_center[n, 0]
                #cluster_center_prev[n, 1] = cluster_center[n, 1]
                for g in range(self.attributes):
                        cluster_center_prev[n, g] = cluster_center[n, g]


                #cluster_center[n, 0] = self.X[c_center, 0]
                #cluster_center[n, 1] = self.X[c_center, 1]
                for g in range(self.attributes):
                        cluster_center[n, g] = self.X[c_center,g]


            #print()
            print("Cluster centers:")
            for q in range(self.clusters):
                #print(cluster_center[q, 0] , cluster_center[q, 1])
                print("cluster-" , q, cluster_center[q])
            
            
            print("-----------------")
            print()
            
            flag = 0
            for n in range(self.clusters):
                for g in range(self.attributes):
                    if(abs(cluster_center_prev[n, g] - cluster_center[n, g]) > epsilon):
                        flag = 1
                        break
                if(flag == 1):
                    break
            if(flag == 0):
                break
            
            
print("Enter the number of clusters:")
y=int(input())
ob = Kmedoid(y)
ob.kmedoid_clustering()
```
